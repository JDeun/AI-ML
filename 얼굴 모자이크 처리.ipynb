```python
#모자이크 처리해보기
import cv2

# Haar Cascade 분류기 로드
face_cascade = cv2.CascadeClassifier(cv2.data.haarcascades + 'haarcascade_frontalface_default.xml')

# 웹캠 열기
cap = cv2.VideoCapture(0)

while True:
    # 웹캠 프레임 읽기
    ret, frame = cap.read()

    # 그레이스케일로 변환
    gray = cv2.cvtColor(frame, cv2.COLOR_BGR2GRAY)

    # 얼굴 탐지
    faces = face_cascade.detectMultiScale(gray, scaleFactor=1.1, minNeighbors=5, minSize=(30, 30))

    # 탐지된 얼굴에 모자이크 처리
    for (x, y, w, h) in faces:
        roi = frame[y:y+h, x:x+w]
        roi = cv2.resize(roi, (10, 10))
        roi = cv2.resize(roi, (w, h), interpolation=cv2.INTER_NEAREST)
        frame[y:y+h, x:x+w] = roi

    # 결과 화면 표시
    cv2.imshow('Face Detection', frame)

    # 'q' 키를 누르면 종료
    if cv2.waitKey(1) & 0xFF == ord(' '):
        break

# 웹캠 및 창 닫기
cap.release()
cv2.destroyAllWindows()
```

roi = frame[y:y+h, x:x+w] : 탐지된 얼굴 영역을 선택합니다.

roi = cv2.resize(roi, (10, 10)) : 탐지된 얼굴 영역을 10x10 크기로 축소합니다.

roi = cv2.resize(roi, (w, h), interpolation=cv2.INTER_NEAREST) : 축소된 영역을 다시 원래 크기로 확대합니다. 이 과정에서 모자이크 효과가 발생합니다.

frame[y:y+h, x:x+w] = roi : 모자이크 처리된 영역을 원래 프레임에 다시 삽입합니다.

```python
import cv2

# Haar Cascade 분류기 로드
face_cascade = cv2.CascadeClassifier(cv2.data.haarcascades + 'haarcascade_frontalface_default.xml')

# 웹캠 열기
cap = cv2.VideoCapture(0)

# 가면 이미지 불러오기
mask_path = r'C:\Users\gadi2\OneDrive\바탕 화면\study file\sticker.png'
mask = cv2.imread(mask_path, cv2.IMREAD_UNCHANGED)

while True:
    # 웹캠 프레임 읽기
    ret, frame = cap.read()

    # 그레이스케일로 변환
    gray = cv2.cvtColor(frame, cv2.COLOR_BGR2GRAY)

    # 얼굴 탐지
    faces = face_cascade.detectMultiScale(gray, scaleFactor=1.1, minNeighbors=5, minSize=(30, 30))

    # 감지된 얼굴에 가면 이미지 적용
    for (x, y, w, h) in faces:
        # 가면 이미지의 크기 조정
        mask_resized = cv2.resize(mask, (w, h))

        # 가면 이미지를 원본 프레임에 적용
        frame[y:y+h, x:x+w] = frame[y:y+h, x:x+w] * (1 - mask_resized[:, :, 3:] / 255.0) + mask_resized[:, :, :3] * (mask_resized[:, :, 3:] / 255.0)

    # 결과 화면 표시
    cv2.imshow('Face Detection with Mask', frame)

    # 'q' 키를 누르면 종료
    if cv2.waitKey(1) & 0xFF == ord('q'):
        break

# 웹캠 및 창 닫기
cap.release()
cv2.destroyAllWindows()
```
