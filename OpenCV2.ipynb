## 이미지 비교

```python
import cv2              # OpenCV 라이브러리 가져오기 pip install opencv-python
import numpy as np      # Numpy 라이브러리 가져오기
import IPython.display
import PIL.Image
import matplotlib.pyplot as plt
```

```python
img1 = cv2.imread("[Dataset] Module 20 images/test0.jpg", cv2.IMREAD_GRAYSCALE)
img2 = cv2.imread("[Dataset] Module 20 images/test1.jpg", cv2.IMREAD_GRAYSCALE)


img1_arr = np.array(img1)
IPython.display.display(PIL.Image.fromarray(img1_arr))


img2_arr = np.array(img2)
IPython.display.display(PIL.Image.fromarray(img2_arr))


diff_img = cv2.absdiff(img1, img2)
IPython.display.display(PIL.Image.fromarray(diff_img))

min_pix_value = 2
ret, thresh = cv2.threshold(diff_img, min_pix_value, 255, cv2.THRESH_BINARY)  #영상을 이진화 처리. 2를 넘으면 255(흰색)으로 변경, 아니면 검은색
contoursArray, hr = cv2.findContours(thresh.copy(), cv2.RETR_EXTERNAL, cv2.CHAIN_APPROX_SIMPLE)[-2:] #윤곽선(컨투어) 검색
```

```python
sensitivity = 5

height, width = img1.shape
width = int(width/2)
height = int(height/2)

if len(contoursArray) > 0 :
    contours = sorted(contoursArray, key=cv2.contourArea, reverse=True)

    # Select the contour with the largest area
    largest_contour = contours[0]
    x, y, w, h = cv2.boundingRect(largest_contour)
    area = w*h

    cv2.rectangle(img2, (x, y), (x+w, y+h), (0, 255, 0), 2)
    IPython.display.display(PIL.Image.fromarray(img2))
    
    if(area > width*height/sensitivity) :
        print('침입자 발생')
        print(area, width, height, width*height/sensitivity)
    else :
        print(area, width, height, width*height/sensitivity)
        print('감지 영역이 기준보다 작습니다')
else : print('이상없음')
```
