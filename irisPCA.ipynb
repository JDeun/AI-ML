```python
import random 
import numpy as np 
import pandas as pd
import matplotlib.pyplot as plt 
from sklearn.neighbors import KNeighborsClassifier
from sklearn.preprocessing import StandardScaler
from IPython.display import display
from sklearn.cluster import KMeans 
from sklearn.decomposition import PCA
from sklearn.metrics import homogeneity_score, completeness_score, \
v_measure_score, adjusted_rand_score, adjusted_mutual_info_score, silhouette_score
import warnings
warnings.filterwarnings('ignore')
```

```python
df = pd.read_csv("./[Dataset]_Module_36_(iris).data",header=None)
names = ["sepal_length", "sepal_width","petal_length", "petal_width", "class"]
df.columns = names
df.head()
```

```python
def Performance_output(data_frame, y_clust, labels, inertia) :
    df = pd.DataFrame({'clust_label': y_clust, 'orig_label': labels})
    ct = pd.crosstab(df['clust_label'], df['orig_label'])
    display(ct)
    print('% 9s' % 'inertia  homo    compl   v-meas   ARI     AMI     silhouette')
    print('%i       %.3f   %.3f   %.3f   %.3f   %.3f    %.3f'
      %(inertia,
      homogeneity_score(labels, y_clust),
      completeness_score(labels, y_clust),
      v_measure_score(labels, y_clust),
      adjusted_rand_score(labels, y_clust),
      adjusted_mutual_info_score(labels, y_clust),
      silhouette_score(data_frame, y_clust, metric='euclidean')))#statistics for clusters
```

K-Means()로 각종 지표값 출력 

```python
Data = df[["sepal_length", "sepal_width","petal_length", "petal_width"]]
labels = df['class']

k = 3
kmeans = KMeans(n_clusters=k, init='random', random_state=0)  # KMeans 객체 생성
kmeans.fit(Data)  # KMeans 모델 훈련
y_clust = kmeans.predict(Data)
Performance_output(Data, y_clust, labels, kmeans.inertia_)
```

```python
def pca_transform(n_comp, data_frame):
    pca = PCA(n_components=n_comp)
    return pca.fit_transform(data_frame)
```

```python
#3차원
Data_reduced = pca_transform(n_comp=3, data_frame=Data)
#print('Reduced Data \n', Data_reduced)

k = 3
kmeans = KMeans(n_clusters=k, init='random', random_state=0)  
kmeans.fit(Data_reduced)  
y_clust = kmeans.predict(Data_reduced)
Performance_output(Data_reduced, y_clust, labels, kmeans.inertia_)
```

```python
#2차원
Data_reduced = pca_transform(n_comp=2, data_frame=Data)
#print('Reduced Data \n', Data_reduced)

k = 3
kmeans = KMeans(n_clusters=k, init='random', random_state=0)  
kmeans.fit(Data_reduced)  
y_clust = kmeans.predict(Data_reduced)
Performance_output(Data_reduced, y_clust, labels, kmeans.inertia_)
```

```python
#1차원
Data_reduced = pca_transform(n_comp=1, data_frame=Data)
#print('Reduced Data \n', Data_reduced)

k = 3
kmeans = KMeans(n_clusters=k, init='random', random_state=0)  
kmeans.fit(Data_reduced)  
y_clust = kmeans.predict(Data_reduced)
Performance_output(Data_reduced, y_clust, labels, kmeans.inertia_)
```

n_comp 매개변수로 분산 비율의 목표를 전달할 수 있습니다. 예를 들어, n_comp=0.95로 설정되면, PCA는 데이터셋의 분산의 95%를 보존하는 주성분 수를 찾습니다. 즉, 반환된 주성분은 원래 데이터의 95% 이상의 분산을 설명할 수 있어야 합니다. 리턴값은 달성된 주성분의 수가 됩니다.

```python
# 분산 비율의 목표치
Data_reduced = pca_transform(n_comp=0.95, data_frame=Data)
print('Reduced Data \n', Data_reduced[:10])

k = 3
kmeans = KMeans(n_clusters=k, init='random', random_state=0)  
kmeans.fit(Data_reduced)  
y_clust = kmeans.predict(Data_reduced)
Performance_output(Data_reduced, y_clust, labels, kmeans.inertia_)
```
