## Scatter plot

산점도(Scatter plot)는 그래프에 점을 사용하여 두 변수의 값을 표시합니다. 이를 통해 두 변수 사이에 관계 또는 상관 관계가 있는지 확인할 수 있습니다.

더 많은 그래프 유형을 알고 있습니까? 아래에 이름을 지정하십시오! 온라인으로 이동하여 matplotlib에서 제공하는 그래프 유형을 찾으십시오.

```python
# 막대그래프 작성 방법

import matplotlib.pyplot as plt

# 데이터 생성
x2 = ['chicken_burger', 'hamburger', 'veggie_burger', 'fish_burger']
y2 = [50, 20, 30, 60]

# 그래프 그리기
plt.bar(x2, y2, color='gray')  # plt.bar에 할당하는 것이 아니라 함수를 호출해야 함

# 데이터 제목과 축 레이블 설정
plt.title('burger bar plot example')
plt.xlabel('burger menu')
plt.ylabel('favorite')

# 그래프 보여주기
plt.show()
```

```python
# 히스토그램 그래프 작성법

import matplotlib.pyplot as plt
import numpy as np # 임의의 숫자 데이터를 생성하기 위함이다. 데이터가 있으면 numpy를 사용하지 않는다.

# 데이터 생성 (정규 분포를 따르는 난수 생성)
data = np.random.randn(1000)

# 히스토그램 그리기
plt.hist(data, bins=30, color = 'red', alpha = 0.7) #bins: 히스토그램의 막대 개수 alpha: 막대의 투명도를 조절

# 그래프 제목과 축 레이블 설정
plt.title('histogram example')
plt.xlabel('value')
plt.ylabel('freaquency')

# 그래프 보여주기
plt.show()
```

```python
# Scikit-Learn에서 유방암 데이터 세트 가져오기
from sklearn.datasets import load_breast_cancer
data = load_breast_cancer()
list(data.target_names)
print(type(data))
print(data)
```

마지막으로, max_depth 인수를 통해 트리의 최대 레벨 수를 지정합니다. 그리고 데이터 셋을 이용하여 적합 함수로 모델을 학습시키고, 플롯 트리를 사용하여 시각화합니다.

# Train, test split

Train, test split은 skit-learn에서 제공하는 기능으로, 그 목적은 데이터 세트를 "훈련 데이터"와 "테스트 데이터"로 분할하는 것입니다. 이 작업을 수행하려는 이유는 모델을 훈련하는 데 사용되지 않은 새 데이터를 기반으로 모델의 성능을 평가하고 데이터 세트가 충분히 큰 경우입니다.

**이 섹션은 보너스입니다! 어떻게 실행하는지 완전히 파악해야 한다는 부담을 갖지 마세요!**

여기서는 이전 예에서 사용된 데이터를 분할합니다.

```python
from sklearn.model_selection import train_test_split  #테스트 데이터와 트레인 데이터를 분리해주는 함수 train_test_split
from sklearn.datasets import load_breast_cancer
data = load_breast_cancer()
# 데이터를 피쳐와 레이블로 분할합니다. 피쳐는 입력이고 레이블은 출력입니다.
X = data.data
y = data.target

# 분할하기 전 데이터의 모양을 시각화하여 볼 수 있습니다.
print(X.shape)
print(y.shape,'\n')

# 다음 변수를 만들고 train_test_split 함수를 사용합니다.
# test size는 test와 train 사이에 데이터가 어떠한 비율로 분할되었는지를 나타내며, 
# 이 경우 0.2는 20%입니다.
X_train, X_test, y_train, y_test = train_test_split(X,y,test_size=0.2) 

# 분할된 데이터의 모양을 확인할 수 있습니다.
print(X_train.shape)
print(X_test.shape,'\n')

print(y_train.shape)
print(y_test.shape)
```

```python
import matplotlib.pyplot as plt
import numpy as np

x = np.linspace(0, 2 * np.pi, 400)
y1 = np.sin(x**2)
y2 = np.cos(x**2)

fig, axs = plt.subplots(2, 2)  # plt.subplot 대신 plt.subplots 사용

axs[0, 0].plot(x, y1)
axs[0, 0].set_title('Axis [0,0]')

axs[0, 1].plot(x, y2, 'tab:orange')  # 서브플롯 위치 수정
axs[0, 1].set_title('Axis [0,1]')

axs[1, 0].plot(x, -y1, 'tab:green')  # 서브플롯 위치 수정
axs[1, 0].set_title('Axis [1,0]')

axs[1, 1].plot(x, -y1, 'tab:red')  # 서브플롯 위치 수정
axs[1, 1].set_title('Axis [1,1]')

plt.tight_layout()  # 서브플롯 간의 간격을 조절하여 보기 좋게 정렬

plt.show()
```

```python
fig, ax1 = plt.subplots()

ax1.set_xlabel('x')
ax1.set_ylabel('sin(x)', color = 'tab:red')
ax1.plot(x, np.sin(x), color = 'tab:red')

ax2 = ax1.twinx()
ax2.set_ylabel('cos(x)', color = 'tab:blue')
ax2.plot(x, np.cos(x), color = 'tab:blue')

plt.show()
```
